```
# -*- coding: utf-8 -*-
"""
Created on Mon Sep 16 10:56:34 2019

@author: cvgq
"""

import pandas as pd
import numpy as np

df1Crash = pd.read_csv('2018_DATA_SA_Crash.csv')
df2Units = pd.read_csv('2018_DATA_SA_Units.csv')

# print(df1Crash.head(20))

# print(df2Units.head(20))

# Round-robin data partitionining function
def rr_partition(data, n):
    # """
    # Perform data partitioning on data

    # Arguments:
    # data -- an input dataset which is a dataframe
    # n -- the number of processors

    # Return:
    # result -- the paritioned subsets of D
    # """
    
    result = [];
    for i in range(n):
        result.append([])
    
    data2 = data.values.tolist()
    
    ### START CODE HERE ###        
    # For each bin, perform the following
    for index, element in enumerate(data2): 
        # Calculate the index of the bin that the current data point will be assigned
        index_bin = (int) (index % n)
        #print(str(index) + ":" + str(element))
        result[index_bin].append(element)
    ### END CODE HERE ###
    
    return result

output = rr_partition(df1Crash, 3)

data = output[0]

import multiprocessing as mp # For multiprocessing

# Linear search function
def linear_search(partition_index, data, key):
    """
    Perform linear search on data for the given key

    Arguments:
    data -- an input dataset which is a list or a numpy array
    key -- an query record

    Return:
    result -- the position of searched record
    """
    
    matched_record = None
    position = -1 # not found position
    resultIndex = []
    record_index = 0
    
    ### START CODE HERE ### 
    for y in data:
        for x in y:
            if x == key: # If x is matched with key
                matched_record = x
                position1 = y.index(x)
                position2 = data.index(y)
                # Get the index of x
                result2 = [partition_index, position2, position1, matched_record]
                resultIndex.append(result2)
                record_index = record_index + 1    
    ### END CODE HERE ###
    
    return resultIndex

# y = linear_search(data, 'ADELAIDE')

def parallel_search_exact_rr_linear(data, query, n_processor):
    results = []
    pool = mp.Pool(processes = n_processor)

    DD = rr_partition(data, n_processor)
    partition_no = 0
    for d in DD:
        rr_result = pool.apply_async(linear_search, [partition_no, d, query])
        output = rr_result.get()
        results.append(output)
        partition_no = partition_no + 1
    
    results2 = list(itertools.chain.from_iterable(results))
    return results2

z_output = parallel_search_exact_rr_linear(df1Crash, 'ADELAIDE', 3 )
```
